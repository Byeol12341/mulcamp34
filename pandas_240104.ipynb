```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회

# day가 Sun인 데이터만 조회하기

# tip의 평균보다 크면서, time이 Dinner인 데이터만 조회
```

```python
tips['total_bill'].median()
```

```python
# total_bill의 중간값 이하인 것만 조회
tips.loc[tips['total_bill'] <= tips['total_bill'].median(), :] # .head(1) 마지막에 추가하면 하나만 나옴
```

```python
# day가 Sun인 데이터만 조회하기
tips.loc[tips['day'] == 'Sun', :]
```

```python
# tip의 평균보다 크면서, time이 Dinner인 데이터만 조회
tips['tip'].mean()
```

```python
tips.loc[tips['tip'] >= tips['tip'].mean(), :].head(1)
```

```python
# tip의 평균보다 크면서, time이 Dinner인 데이터만 조회
# AND 조건, A이고 B이다. A이면서 B이다.
# OR 조건, A이거나 또는 B이다.
# tips.loc[(조건1) & (조건2), :]

tips.loc[(tips['tip'] >= tips['tip'].mean()) & (tips['time'] == 'Dinner'), :]
```

```python
# tips.loc[(조건1) & (조건2), :]
# AND 조건 / |, OR 조건
# 다중 조건 예시
tips.loc[(tips['tip'] >= tips['tip'].mean()) &
         (tips['time'] == 'Dinner') &
         (tips['size'] == 3) &
         (tips['sex'] == 'Male')&
         (tips['day'] == 'Sat'),
:]
```
